# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data

```python
df = pd.read_csv("../data/processed/cars_cleaned.csv")

df.head()
```

Basic statistics

```python
df.describe()
```

Plot Price Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=50)
plt.title("Price Distribution")
plt.show()
```

Check Correlation Between Numeric Features

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["price","mileage","engine_capacity","car_age"]].corr(), 
            annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Car Age vs Price

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="car_age", y="price", data=df)
plt.title("Car Age vs Price")
plt.show()
```

Mileage vs Price

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="mileage", y="price", data=df)
plt.title("Mileage vs Price")
plt.show()
```

Brand Effect on Price

```python
plt.figure(figsize=(12,6))
top_brands = df["brand"].value_counts().head(10).index
sns.boxplot(data=df[df["brand"].isin(top_brands)],
            x="brand", y="price")
plt.xticks(rotation=45)
plt.title("Price Distribution by Top Brands")
plt.show()
```
